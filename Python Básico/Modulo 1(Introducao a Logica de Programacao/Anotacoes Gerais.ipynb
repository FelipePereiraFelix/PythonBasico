# Modulo 1: Introdução á lógica de Programção

Na iformática, uma instrução(comando) indica a um computador uma ação elementar a ser executada.

Para que lógica funcione a uma obrigação de seguir uma Ordem Adequada, um exemplo seria o de beber um refigerante
em lata, sendo as seguintes lógicas:

>> 1. Pegar a lata
>> 2. Abrir a tampa
>> 3. Beber do conteúdo

A máquina desconhece conceitos que temos como claros, ou óbvios em nossa mente. Por isto, há uma necessidade de 
descrever passo por passo para que haja uma **sequência lógica na 
programação** e, assim, o computador possa executar todas as instruções necessárias para uma determinada tarefa. 

## Por que a Lógica de Programação é tão Importante?

Ela é o processo de procurar pensar na **sequência** em que o computador vai executar as tarefas. 

Com isto, segue uma lógica de programação para que imaginemos como as 
**ações serão executadas**, assim se constuindo um **algoritmo** para que possamos entender os problemas e soluções dele. 

### O que seria o Algoritmo? 

O algoritmo é uma **Sequência finita** de passo que levam a execução de uma tarefa. 
Uma forma de pensar seria que o algoritmo é como uma receita, ou seja, apresenta uma sequência de instruções para um fim específico. 

Essas instruções devem ser descritas de maneira simples e objetivas, poi caso haja uma instrução descrita errada o código não será executado. 

Um **Algoritmo deve** sempre:

- ter um Ponto Inicial e Ponto Final
- Não pode ser ambíguo, ou seja, não pode ter dupla interpretação.
- Deve receber dados exteros para ser capaz de retonar resultados. 
- Deve ter estapas alcançáveis. 

Um **Algoritmo** também deve seguir uma composição de algoritmo:

- Entrada: São os dados de entrada do algoritmo
- Processamento: São os procedimento utilizadoss para chegar ao resultado final. 
- Saida: São os dados ja processados, ou seja, o resultado da programação.

As formas de representação são as:

- Descrição Narrativa: Descrever de forma clara o que está ali, mesmo que a pessoa não conheça sobre algoritmos.

Ex: 

>> Passo 1: Receber os números que serão multiplicados; 
>> Passo 2: Multiplicar os números;
>> Passo 3: Exibir o resultado da multiplicação;

- Fluxograma: Atreavés de gráficos para repressentar os algoritmos.

Ex:
>> Incio 
>> Leia num 1
>> Leia num 2
>> Resultado = Num1 * Num 2
>> Resultado 
>> Fim

- Pseudocódigo: Seria a base do Portugol ou Português Estruturado, porta principal de entrada para linguagem de programação. Usando comandos em português para demonstrar algoritmos.

Ex: 

>> Declare num1, num2, resultado;
>> Leia num1, num2;
>> Resultado = num1+num2;
>> Escreva Resultado.

**Operadores Lógicos**

- E AND
- OU OR

Exemplos: 

- E/AND: O resultado de uma expressão lógica usando o operador lógico AND é verdadeira, **somente se todas condições forem verdadeiras.

    (A>B) AND (A>C) = VERDADEIRO

Ex: (3>5) OR (8>1) = VERDADEIRO
    (A>B) OR (A>C) = VERDADEIRO

- NÃO/NOT: O resultado de uma expressão lógica usando o operador lógico NOT, inverte o valor da expressão ou condição, ou seja, **se verdadeira inverte para falsa e vice-versa**. 

Ex: NOT(3>5) = VERDADEIRO
    NOT(A>B) = FALSO

### Estrutura de decisão

A estrutura de decisão, também conhecida como **estrutura condicional**, tem como finalidade comparar e efetuar um desvio de processamento do programa. 

Caso a condição seja verdadeira, as instruções determinadas após a instrução **if** serão executadas. 

Se for necessário executar mais de uma intrução para a condição verdadeira, elas deverão estar escritas dentro de um bloco, ou seja, identadas corretamente. 

### Estrutura de decisão simples if(se)

A estrutura de condição **if** permite avaliar uma expressão e, se o resultado for verdadeiro, será executada uma determinada ação. 

No código abaixo, temos a verificação da variável idade. Caso o valor da variável seja maior que 18, a condição if é verdadeira, portanto, a mensagem "maior idade" será exibida. 

idade = int (input("Digite a idade da pessoa: "))

if dade > 18
   
   print("maior idade")


Caso a condição seja falsa (idade menor que 18), nenhuma instrução será realizada.

  print("Menor idade")

### Estruturas de Repetição

As estruturas de repetição, também conhecidas como laços de repetição, permitem que um conjunto de instruções eja executado, até que uma determinada condição seja verdadeira. 

As estruturas de repetição em Python são:

### Estrutura For(para)

Normalmente utilizamo a estrutura **For** quando sabemos quantas vezes o laço de programação deverá ser executado. 

É a estrutura mais utilizada na linguagem Python. Ela aceita sequências estáticas e geradas por **iteradores**. 

**Você sabe** o que são iteradores?

**Iteradores** são estruturas que permitem acesso a vários itens de uma coleção de elementos, de forma sequencial.

Durante a execução de uma estrutura **For**, o contador aponta para um elemento de uma determinada sequência. A cada iteração, o contador é atualizado para que o laço **For** processe os elementos correspondentes. 

Sintaxe da estrutura For:

For{Referência}
in{sequência}: {bloco de código}

A estrtura **While(enquanto)**, executa um determinado conjunto de instruções, enquanto a condição verificada no início permanecer verdadeira.

# Módulo 5: Funções Práticas de programação com funções

### Funções 

Funções ssão estruturas que possibilitam a separação da programação em blocos.

A implementação de funções em programas é fundamental, pois tem o objetivo de otimizar o código-fonte e evitar a replicação do código, ou seja, definimos as funções uma única vez e sempre que necessário podemos utilizá-las. 

### Sintaxe de uma função

A declaração de uma função é dividida em três partes: nome, parâmetros e corpo.

def **nome_da_função** (**parâmetros**)

      <instruções>

**return** "valor do retorno".


Onde:

>> 1. def : A palavra **def** determina o inicio da função.

>> 2. Parâmetros: São informações que a função pode receber para o seu processamento. Os parâmetros podem existir ou não

>> 3. Corpo da função: É o local que é aplicada a sequência de instruçõe, como entrada, processamento e/ou saída.

>> 4. Return: Deve er utilizado quando houver a necessidade de retonar alguma informação para a ação da função.

### Funções para manipução de textos

Por meio da linguagem Python, podemo manipular dados em um arquivo texto, mostrando operações básicas de leitura e escrita.

Esse arquivo texto é conhecido como **arquivo sequencial**, porque a leitrua tem de ser feita de forma sequencial, do início ao fim do arquivo.

Para trabalharmo com arquivos texto na linguagem Python utilizamoss funções.

### Função open()

A função **open()** é utilizada pra abertura dos arquivos
             
             Sua sintaxe é:

**arquivo=open('arquivo.txt', 'w')**

A função **open()**, após declaração da variável que receberá a função, necessita de dois parâmetros: Primeiramente o nome do arquivo e, depois, o modo como estamos abrindo esse arquivo.

Na sintaxe apresentada acima foi utilizado o '**W**' para fazer a escrita em um arquivo.

Caso o arquivo não existe nesse modo, o código criará um arquivo com nome escrito no primeiro parâmetro.

**!!Importante**: Reescrever o arquivo

Ao utilizar o 'w' em uma arquivo ja existente, ele reescreverá todo o conteúdo do arquivo, fazendo com que todos os dados que estavam nele sejam apagados.

### Função write()

A função **write()** é utilizada para gravar informações em um arquivo existente.

      Sua sintaxe é:

**arquivo.write('Curso Python n')**

**arquivo.write('Aula Prática')

Na função, adicionamos o nome do arquivo e, logo após o símbolo do ponto final, fazemo a chamada da função **write**. Em seguida, adicionamos o texto que deverá ser gravado entre aspas simples.

### Função close()

A função **close()** é muito importante para encerrar o arquivo após sua utilização.

**Atenção**: Nunca abra o arquivo com a função open e depois faça de novo, sem antes fechar a instância anterior.

       Sua sintaxe é:

**arquivo.close()**

Um dos motivos da necessidade da função **close()** é que se tentarmos escrever em um arquivo e não o fecharmos depois de terminar a escrita, ass informações não chegarão ao arquivo e nada será escrito.

>> Na prática: va para o arquivo "Gravar.py"

### Função read()

A função **read()** realiza a leitrua de todo o conteúdo do arquivo 

    Sua sintaxe é:

**leitura=open('aquivo.txt','r')**

**print leitura.read()**

**leitura.close()**

Utilizamos o 'r' que representa que o arquivo está sendo aberto em modo leitura.

Desta forma, não é possível modificar os dados contidos no arquivo. 

**!! Importante**

**Nome do arquivo**

Ao utilizar a função **read()**, o nome do arquivo deve ser:

1) Uma *string* com caminho completo(por exemplo, C Documentos texte text)

                        ou

2) O caminho em relação ao diretório atual(nomes txt do arquivo que se deseja abrir).

### Modos de leitura de arquivos

Além dos comandos vistos até aqui (**w** e **r**), também temos disponíveis outros comandos que podemos utilizar nos parâmetros. 

Vamo conhecê-los?

>> Modo | Descrição

>> r | Abre o arquivo somente para leitura. O arquivo deve ja existir.

>> r+ | Abre o arquivo para leitura e escrita. O arquivo deve já exisitir. A escrita começa a partir da primeira linha e, caso exista algo escrito no arquivo, as linhas serão reescritas, conforme formos escrevendo.

>> w+ | Abre o arquivo somente para escrita, no início do arquivo. Apagará o conteúdo do arquivo se ele já existir. Criará um arquivo novo se não existir.

>> a | Abre o arquivo para escrita no final do arquivo. Não apaga o conteúdo preexistente.

>> a+ | Abre o arquivo para escrita final do arquivo e leitura.
